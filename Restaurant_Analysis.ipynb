```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.style.use('fivethirtyeight')

nonVeg_restos = ['KFC', 'Mc.Donald', 'Sigdi', 'Subway', 'Kebab and Curries']

nonVeg_resto_price = [1200, 750, 950, 650, 700]    

fig = plt.figure(figsize = (7,3), dpi= 130)

plt.barh(nonVeg_restos,nonVeg_resto_price, edgecolor = 'black', linewidth = 2)

nonVeg_restos.reverse()
nonVeg_resto_price.reverse()

plt.title('NonVeg Restos Price Analysis')
plt.ylabel('Restaurants Name')
plt.xlabel('Price')

plt.tight_layout()
```

# Advance Non-Veg_Resto's Price Analysis For to Know which one is Cheap and Affordable!

```python
# matplotlib and numpy are already imported above!

plt.style.use('fivethirtyeight')

fig = plt.figure(figsize = (7,4), dpi=100)

Resto_Name_List = []
Resto_Price_List = []

while input("\n\t\t\t\tTo Add Restaurant Name, Type 'Y' else Type 'N': ") == 'Y':
    
    # Resto Name Input
    Name_Input = input('\n\n\t\tEnter the Name of the Restaurant: ').strip()    # KFC
    Resto_Name_List.append(Name_Input)                                  # Resto_Name_List = ['KFC', 'McD']
    
    # Price Input
    Price_Input = input('\n\t\tEnter the Prices of items to get the mean: ').split(' ')    
    # ['20', '30', '40'] -> [20, 30, 40] below...
    
    Float_Conversion = list(map(float, Price_Input))
    Mean_Price = np.mean(Float_Conversion)
    Resto_Price_List.append(Mean_Price)
    
    print("\n\t\tMean Price of '",Name_Input, "' Restaurant is: ", Mean_Price)
    print("\n\t--------------------------------------------------------------------------------------")
    
print('\n\t\tList of Non-Veg Restaurants Being Added By You: ', Resto_Name_List)
print('\n\t\tList of Non-Veg Restaurants Mean Prices: ', Resto_Price_List)

plt.barh(Resto_Name_List, Resto_Price_List, edgecolor = 'black', lw = 2)

Resto_Name_List.reverse()
Resto_Price_List.reverse()

plt.title('NonVeg Restos Price Analysis')
plt.ylabel('Restaurants Name')
plt.xlabel('Price')

plt.tight_layout()
```
